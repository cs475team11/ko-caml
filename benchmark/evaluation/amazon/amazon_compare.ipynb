````python
import pandas as pd
import json

# Prepare the result data

# Helper function to parse the answer
def parse_answer_kor(ans):
    return json.loads(ans.strip("```").lstrip("json").replace("\n", "").replace(" ", "")).get("가장_적합한_HS_코드", "").rjust(6, "0")

def parse_answer_eng(ans):
    return json.loads(ans.strip("```").lstrip("json").replace("\n", "").replace(" ", "")).get("BestHSCode", "").rjust(6, "0")

# Load the CSV files
result_eng = "../results/12072329_amazon_result_eng.csv"
result_kor = "../results/12072329_amazon_result_kor.csv"

# Read the CSV files into DataFrames
result_eng = pd.read_csv(result_eng)
result_kor = pd.read_csv(result_kor)

result_eng['Answer'] = result_eng['Answer'].apply(parse_answer_eng)
result_kor['정답'] = result_kor['정답'].apply(parse_answer_kor)
````

````python
import pandas as pd
# sample_result = pd.read_csv("../results/12201616_sample_amazon_result_eng.csv")
sample_result = pd.read_csv("../results/12072329_sample_amazon_result_kor.csv")
````

````python
sample_result['Answer'] = sample_result['Answer'].apply(lambda x: str(x).rjust(6, "0"))
sample_result
````

````python
ground_truth = pd.read_csv("amazon_gt.csv")
ground_truth['Ground Truth'] = ground_truth['Ground Truth'].apply(lambda x: str(x).rjust(6, "0"))
ground_truth
````

````python
merged_df = pd.concat([ground_truth, sample_result], axis=1)
````

````python
merged_df = pd.read_csv("sample_amazon_merged_kor.csv")
merged_df.drop(columns=['pn'], inplace=True)
merged_df.to_csv("sample_amazon_merged_kor.csv", index=False)
````

````python
merged_df['kor_Match_2'] = [0] * 19
merged_df['kor_Match_4'] = [0] * 19
merged_df['kor_Match_6'] = [0] * 19
````

````python
merged_df
````

````python
merged_df.loc[0, 'Ground Truth']
````

````python
for i, row in merged_df.iterrows():
    # Create a dictionary from the row
    gt = str(merged_df.loc[i, "Ground Truth"]).rjust(6, "0")
    kor = str(merged_df.loc[i, "Answer"]).rjust(6, "0")
    merged_df.loc[i, "kor_Match_2"] = kor[:2] == gt[:2]
    merged_df.loc[i, "kor_Match_4"] = kor[:4] == gt[:4]
    merged_df.loc[i, "kor_Match_6"] = kor == gt
````

````python
merged_df.to_csv("sample_amazon_merged_w_result_kor.csv", index=False)
````
